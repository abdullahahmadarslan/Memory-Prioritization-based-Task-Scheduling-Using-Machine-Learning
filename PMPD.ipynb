```python
import pandas as pd

# Load the extracted CSV
columns = ["task_name", "instance_num", "job_name", "task_type", "status", "start_time", "end_time", "plan_cpu", "plan_mem"]
task_df = pd.read_csv("extracted_batch_task/batch_task.csv", names=columns)

# Sample 5000 from each status
sampled_df = task_df.groupby('status', group_keys=False).apply(lambda x: x.sample(n=10000, random_state=42)).reset_index(drop=True)
```

```python
# STEP 2: Stream VM Requests with Kafka

from kafka import KafkaProducer
import json
import pandas as pd

# Set up Kafka producer
producer = KafkaProducer(bootstrap_servers='localhost:9092',
                         value_serializer=lambda v: json.dumps(v).encode('utf-8'))

for _, row in sampled_df.iterrows():
    task = {
        "task_name": row.task_name,
        "job_name": row.job_name,
        "cpu": row.plan_cpu,
        "mem": row.plan_mem,
        "start_time": int(row.start_time),
        "end_time": int(row.end_time)
    }
    producer.send("tasks", task)

producer.flush()  # Make sure all messages are sent
```

```python
# STEP 3: Read Kafka as Batch in Spark

from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StringType, DoubleType, LongType
from pyspark.sql.functions import from_json, col
import pandas as pd

spark = SparkSession.builder \
    .appName("TaskScheduler") \
    .config("spark.jars.packages", "org.apache.spark:spark-sql-kafka-0-10_2.12:3.4.1") \
    .getOrCreate()

# Spark reads Kafka topic "tasks" in parallel across partitions
tasks_batch = spark.read.format("kafka") \
    .option("kafka.bootstrap.servers", "localhost:9092") \
    .option("subscribe", "tasks") \
    .load()

# Parse Kafka value column (JSON)
parsed_batch = tasks_batch.selectExpr("CAST(value AS STRING) as json_data") 

task_schema = StructType() \
    .add("task_name", StringType()) \
    .add("job_name", StringType()) \
    .add("cpu", DoubleType()) \
    .add("mem", DoubleType()) \
    .add("start_time", LongType()) \
    .add("end_time", LongType())

# Spark's parallel columnar processing
final_batch = parsed_batch.select(from_json(col("json_data"), task_schema).alias("data")).select("data.*") 

# Convert to Pandas for scheduling
batch_df = final_batch.toPandas() # High memory usage

batch_df['start_time'] = pd.to_numeric(batch_df['start_time'], errors='coerce')
batch_df['end_time'] = pd.to_numeric(batch_df['end_time'], errors='coerce')

# Drop any rows where either start_time or end_time is missing
batch_df = batch_df.dropna(subset=['start_time', 'end_time'])

print("Kafka batch successfully read into Spark and parsed")
```

```python
# STEP 4: Cluster PMs Using Machine Usage

from pyspark.ml.clustering import KMeans
from pyspark.ml.feature import VectorAssembler
import pandas as pd

# Load machine usage file
machine_df = pd.read_csv("extracted_machine_usage/machine_usage.csv", names=[
    "machine_id", "time_stamp", "cpu_util_percent", "mem_util_percent",
    "mem_gps", "mkpi", "net_in", "net_out", "disk_io_percent"
],    nrows=100000  # Reads only the first 10,000 rows
)

# Convert to Spark DataFrame
machine_spark_df = spark.createDataFrame(machine_df)   

assembler = VectorAssembler(
    inputCols=["cpu_util_percent", "mem_util_percent"],
    outputCol="features"
)

vector_data = assembler.transform(machine_spark_df) # Vectorizes data in parallel

# Apply KMeans clustering (k=3 clusters)
kmeans = KMeans(k=3, seed=1)
model = kmeans.fit(vector_data)  # 	KMeans clusters data in parallel
clustered = model.transform(vector_data)  # Adds prediction column in parallel

# Extract cluster centers dynamically
cluster_centers = model.clusterCenters()

# Convert to Pandas for scheduling
clustered_pd = clustered.select("machine_id", "cpu_util_percent", "mem_util_percent", "prediction").toPandas()

print("PM clustering complete")
```

```python
# STEP 5: PMPD Scheduler Preparation

# Load machine_meta file to get total CPU/memory
meta_columns = ["machine_id", "time_stamp", "failure_domain_1", "failure_domain_2", "cpu_num", "mem_size", "status"]
machine_meta = pd.read_csv("extracted_machine_meta/machine_meta.csv", names=meta_columns)

# Merge usage data with meta data
pm_df = pd.merge(clustered_pd, machine_meta[["machine_id", "cpu_num", "mem_size"]], on="machine_id")

# Calculate available free CPU/mem
pm_df["cpu_free"] = pm_df["cpu_num"] * (100 - pm_df["cpu_util_percent"]) / 100
pm_df["mem_free"] = pm_df["mem_size"] * (100 - pm_df["mem_util_percent"]) / 100

# Create PM dictionary for scheduling
pm_list = pm_df.to_dict("records")

print(" PM resource info prepared")
```

```python
import pandas as pd

def save_pm_status():
    df = pd.DataFrame(pm_list)
    df.to_csv("pm_status.csv", index=False)
```

```python


# PART 6: PMPD Scheduling with resource release simulation

import time
import threading
from collections import Counter
import numpy as np

assignments = []
running_tasks = []
lock = threading.Lock()

# Dynamic scheduler function with inverted cluster logic and global fallback
def pmpd_scheduler(task, pm_list, cluster_centers):
    # Compute task ratio
    task_ratio = task['cpu'] / task['mem'] if task['mem'] != 0 else float('inf')

    # Compute cluster ratios dynamically
    cluster_ratios = np.array([cpu / mem if mem != 0 else float('inf') for (cpu, mem) in cluster_centers])
    sorted_ratios = np.sort(cluster_ratios)

    # Intelligent dynamic thresholds (percentile based)
    low_threshold = np.percentile(sorted_ratios, 33)
    high_threshold = np.percentile(sorted_ratios, 66)

    # Inverted cluster assignment logic
    if task_ratio >= high_threshold:
        target_cluster = np.argmin(cluster_ratios)  # Assign CPU-heavy tasks to less CPU-heavy cluster
    elif task_ratio <= low_threshold:
        target_cluster = np.argmax(cluster_ratios)  # Assign MEM-heavy tasks to less MEM-heavy cluster
    else:
        target_cluster = np.argmin([abs(task_ratio - ratio) for ratio in cluster_ratios])  # Balanced

    # First: try PMs inside assigned cluster
    candidate_pms = [pm for pm in pm_list if pm['prediction'] == target_cluster]
    pm_sorted = sorted(candidate_pms, key=lambda pm: pm['mem_free'], reverse=True)

    scheduled = False

    for pm in pm_sorted:
        cpu_ok = pm['cpu_free'] >= task['cpu']
        mem_ok = pm['mem_free'] >= task['mem']
        
        if cpu_ok and mem_ok:
            pm['cpu_free'] -= task['cpu']
            pm['mem_free'] -= task['mem']
            
            duration = task['end_time'] - task['start_time']
            completion_time = time.time() + duration

            with lock:
                running_tasks.append((task, pm, completion_time))
                assignments.append({"task": task['task_name'], "pm": pm['machine_id']})

            save_pm_status()
            print(f" Task {task['task_name']} assigned to PM {pm['machine_id']} (Cluster {target_cluster}) for {duration} seconds")
            scheduled = True
            break

    # Fallback to global PM pool
    if not scheduled:
        print(f" Task {task['task_name']} not scheduled in Cluster {target_cluster}. Trying global fallback.")

        fallback_pms = sorted(pm_list, key=lambda pm: pm['mem_free'], reverse=True)
        for pm in fallback_pms:
            cpu_ok = pm['cpu_free'] >= task['cpu']
            mem_ok = pm['mem_free'] >= task['mem']
            
            if cpu_ok and mem_ok:
                pm['cpu_free'] -= task['cpu']
                pm['mem_free'] -= task['mem']
                
                duration = task['end_time'] - task['start_time']
                completion_time = time.time() + duration

                with lock:
                    running_tasks.append((task, pm, completion_time))
                    assignments.append({"task": task['task_name'], "pm": pm['machine_id']})

                save_pm_status()
                #  Here we print also the cluster where PM belongs
                print(f" Task {task['task_name']} assigned after fallback to PM {pm['machine_id']} (Cluster {pm['prediction']}) for {duration} seconds")
                scheduled = True
                break

    # If still not scheduled  full failure reason
    if not scheduled:
        reasons = []
        for pm in pm_list:
            cpu_shortage = pm['cpu_free'] < task['cpu']
            mem_shortage = pm['mem_free'] < task['mem']
            if cpu_shortage and mem_shortage:
                reasons.append("CPU & Memory insufficient")
            elif cpu_shortage:
                reasons.append("CPU insufficient")
            elif mem_shortage:
                reasons.append("Memory insufficient")

        if reasons:
            common_reason = Counter(reasons).most_common(1)[0][0]
            print(f" Task {task['task_name']} could not be scheduled. Reason: {common_reason}")
        else:
            print(f" Task {task['task_name']} could not be scheduled. Reason: Unknown resource issue.")


# Background monitoring 

def monitor_resources():
    while True:
        now = time.time()
        with lock:
            tasks_to_remove = []
            for t in running_tasks:
                task, pm, completion_time = t
                if now >= completion_time:
                    pm['cpu_free'] += task['cpu']
                    pm['mem_free'] += task['mem']
                    tasks_to_remove.append(t)
                    print(f" Task {task['task_name']} completed and resources released from PM {pm['machine_id']}")
                    save_pm_status()

            for t in tasks_to_remove:
                running_tasks.remove(t)
        time.sleep(1)

monitor_thread = threading.Thread(target=monitor_resources)
monitor_thread.daemon = True
monitor_thread.start()


# Start scheduling incoming tasks

for _, row in batch_df.iterrows():
    task = {
        "task_name": row['task_name'],
        "cpu": row['cpu'],
        "mem": row['mem'],
        "start_time": row['start_time'],
        "end_time": row['end_time']
    }
    pmpd_scheduler(task, pm_list, cluster_centers)

print(" Real-time Intelligent PMPD hybrid scheduling started ")
```
